```python
# counts of cumulative US deaths as of May 4, 2020
# source: https://twitter.com/ryanstruyk/status/1257504253150715904

# cumulative deaths
cd = [
    6, 
    26, 
    86, 
    541, 
    3_003, 
    10_986, 
    23_649, 
    42_295, 
    56_255, 
    68_922, 
]
```

```python
# weeklydeaths 
wd = []

# weekly growth rates
wgr = []

previous_weekly_new_deaths = 0

for weeknum, weekly_cumulative_deaths in enumerate(cd):
    if weeknum == 0 or previous_weekly_new_deaths == 0:
        new_deaths = weekly_cumulative_deaths
        growth_rate = 1
    else:
        new_deaths = weekly_cumulative_deaths - cd[weeknum - 1]
        change_in_deaths = new_deaths - previous_weekly_new_deaths
        growth_rate = change_in_deaths / previous_weekly_new_deaths
    wd.append(new_deaths)
    previous_weekly_new_deaths = new_deaths
    wgr.append(growth_rate)
    
    
# print(f"weekly new deaths:\n{wd}\n")
# print(f"weekly death growth rate:\n{wgr}")
```

```python
import matplotlib.pyplot as plt
plt.plot(cd)
plt.title("weekly cumulative deaths")
plt.xlabel("week number")
plt.show()

plt.plot(wd)
plt.title("weekly new deaths")
plt.xlabel("week number")
plt.show()

plt.plot(wgr)
plt.title("weekly new deaths growth rate")
plt.xlabel("week number")
plt.ylabel("growth rate (in hundred percents)")
plt.show()
```
